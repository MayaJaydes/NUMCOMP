```python
#Swim, bike, run
sports = np.array([1500, 40000, 10000])

Mary = 1
John = 2
Peter = 3
Mafalda = 4 
Paul = 5
Lionel = 6

athletes_times = np.array([[1, 1.204, 6.6, 0.32], [2, 1.212, 6.5, 0.36],
                         [3, 1.14, 6.6, 0.38], [4, 1.12, 6.2, 0.38],
                         [5, 1.24, 5.6, 0.4], [6, 1.201, 6, 0.3]])

#Making sure array shape is (6,4)
print(athletes_times.shape)
```

```python
def triathlon_times(sports, athletes_times):
    """Function takes two arrays to return an informative message about which participant will finish the triathlon first and who will finish last with both participant's expected time to completion (hours).
    Inputs: array containing distances of sports, array containing athlete's IDs as well as their average speed per sport
    Outputs: Names of participant who will finish first and participant who will finish last and time to their completion in hours"""
    
    
    #Finding fastest/slowest time
    
    timess = []
    
    for i in range(len(athletes_times)):
        timeswim = sports[0] / athletes_times[i, 1] 
        timebike = sports[1] / athletes_times[i, 2]
        timerun = sports[2] / athletes_times[i, 3]
        totalsec = timeswim + timebike + timerun
        timess.append(totalsec)
        
        
    timeh = np.round((np.array(timess) / 3600),1)
    
    fastest_time = min(timeh)
    slowest_time = max(timeh)
    
    min_in_list = np.where(timeh == fastest_time)
    max_in_list = np.where(timeh == slowest_time)
    
    
    firstplace = int(athletes_times[min_in_list, 0])
    lastplace = int(athletes_times[max_in_list, 0])
    
    
    return print('Athlete', firstplace, 'will finish first with a time of about', fastest_time , 'hours.','Athlete', lastplace, 'will finish last with a time of about', slowest_time, 'hours.'  )
        
```

```python
triathlon_times(sports, athletes_times)
```

```python
sports = {'Swim' : 1500, 
          'Bike' : 40000, 
          'Run' : 10000}

athletes_times = {'Mary' : [1.204, 6.6, 0.32], 
                  'John' : [1.212, 6.5, 0.36],
                  'Peter' : [1.14, 6.6, 0.38], 
                  'Mafalda' : [1.12, 6.2, 0.38],
                  'Paul' : [1.24, 5.6, 0.4], 
                  'Lionel' : [1.201, 6, 0.3]}
```

```python
def triathlon_time_dict(sports, athletes_times):
    """Function takes two dictionaries and returns an informative message about which participant will finish the triathlon first and who will finish last with both participant's expected time to completion (hours).
    Inputs: array containing distances of sports, array containing athlete's IDs as well as their average speed per sport
    Outputs: Names of participant who will finish first and participant who will finish last and time to their completion in hours"""
   
    timeh = {}
   
    for keys in athletes_times.keys():
        distances = np.array([sports['Swim'], sports['Bike'], sports['Run']])
        timeeach = distances/athletes_times.get(keys)
        totalhr = np.round((sum(timeeach) / 3600), 1)
        timeh.update({keys : totalhr})
        
    fastest_time = min(timeh.values())
    slowest_time = max(timeh.values())
    
    for keys in timeh.keys():
        if timeh[keys] == fastest_time:
            firstplace = keys
        if timeh[keys] == slowest_time:
            lastplace = keys

    return print('Athlete', firstplace, 'will finish first with a time of about', fastest_time , 'hours.','Athlete', lastplace, 'will finish last with a time of about', slowest_time, 'hours.'  )
    
    
```

```python
triathlon_time_dict(sports, athletes_times)
```
