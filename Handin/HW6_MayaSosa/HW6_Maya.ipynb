```python
def Quadratic_Formula(a, b, c):
    """Using coefficients a, b, and c from quadratic function in the form 
    ax + by + c = 0, returns a list containing both roots
    Inputs: a, b, c (coefficients) 
    Outputs: Roots 1 and 2"""
    
    r_1 = (-b + np.sqrt((b**2)-(4*a*c)))/(2 * a)
    r_2 = (-b - np.sqrt((b**2)-(4*a*c)))/(2 * a)
    return [r_1, r_2]
    
```

```python
r = Quadratic_Formula(1, -1, -12)
print('The roots of quadratic function x^2-x-12=0 are', r)
```
